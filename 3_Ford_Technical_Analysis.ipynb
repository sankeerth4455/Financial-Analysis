```python
# Importing relevant libraries
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import ta

plt.style.use("seaborn")
plt.rcParams["figure.figsize"] = [14, 8]
```

```python
#Downloading KnowBe4 stock data of January 2022

df = yf.download("F", start = "2021-01-01", end = "2021-12-31")
df.head()
```

```python
#Plotting stock prices
df['Adj Close'].plot(title='Ford stock chart in 2021')
```

```python
#instantiate the indicator
ta_bbands = ta.volatility.BollingerBands(close=df['Adj Close'], window=21, window_dev=1)

#Copy as we will be adding columns to it temporarily
ta_df=df.copy()

#Adding Bollinger Bands
ta_df['bb_ma'] = ta_bbands.bollinger_mavg()
ta_df['bb_high'] = ta_bbands.bollinger_hband()
ta_df['bb_low'] = ta_bbands.bollinger_lband()

ta_df.head()
```

```python
ta_df[['Adj Close','bb_ma','bb_high','bb_low']].plot(title= "Technical Analysis of Ford stock using Bollinger Bands");
```
